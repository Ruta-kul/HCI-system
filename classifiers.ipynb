```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
features = pd.read_csv("points.csv",encoding="utf-8-sig")
x = features.iloc[:,0:2]
y = features.iloc[:,2]

print(features.shape)
```

```python

# min-max normalization
def min_max_normalize(val, val_min, val_max):
    return float((val - val_min)/(val_max - val_min))
```

```python
# for feature in x.columns:
#     f_min = x[feature].min()
#     f_max = x[feature].max()
#     for i in range(72):
#         x[feature][i] = min_max_normalize(x[feature][i], f_min, f_max)    

    

x_train = x[:525]
x_test = x[525:]
y_train = y[:525]
y_test = y[525:]






print(x_test)
print(y_test)
```

```python
#Import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier

#Create KNN Classifier
knn = KNeighborsClassifier(n_neighbors=4)

#Train the model using the training sets
knn.fit(x_train, y_train)

#Predict the response for test dataset
y_pred = knn.predict(x_test)

# print(y_pred)
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
knn_cross = KNeighborsClassifier(n_neighbors=6)

x = features.iloc[:,0:2]
y = features.iloc[:,2]




kf = KFold(n_splits=10,random_state=42, shuffle=False)

print(x.loc[20])
scores = []

for train_index, test_index in kf.split(x):
    print("Train Index: ", train_index)
    print("Test Index: ", test_index)
    x_train, x_test, y_train, y_test = x.loc[train_index], x.loc[test_index], y[train_index], y[test_index]
    knn_cross.fit(x_train, y_train)
    y_pred = knn_cross.predict(x_test)
    scores.append(accuracy_score(y_pred, y_test))
    
```

```python

np.mean(scores)
```

```python
from sklearn import tree, metrics
```

```python
dtree = tree.DecisionTreeClassifier(criterion='entropy', random_state=0)
dtree.fit(x_train, y_train)
```

```python
y_pred = dtree.predict(x_test)
# how did our model perform?
count_misclassified = (y_test != y_pred).sum()
print('Misclassified samples: {}'.format(count_misclassified))
accuracy = metrics.accuracy_score(y_test, y_pred)
print('Accuracy: {:.2f}'.format(accuracy))
```

```python
from sklearn.ensemble import RandomForestClassifier

kf3 = KFold(n_splits=10,random_state=42, shuffle=False)
scores1 = []




# Create the model with 100 trees
model = RandomForestClassifier(n_estimators=50, 
                               bootstrap = True)
# Fit on training data

for train_index, test_index in kf3.split(x):
    print("Train Index: ", train_index)
    print("Test Index: ", test_index)
    x_train, x_test, y_train, y_test = x.loc[train_index], x.loc[test_index], y[train_index], y[test_index]
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    scores1.append(accuracy_score(y_pred, y_test))
```
